#### Glass Flows Reward Guidance

Reward guidance has the form

$u = u + c \nabla r(x)$

For glass flows, the reward guidance is applied at each inner glass flow time step, to the glass denoiser, and the image latent is transformed with $y = \alpha_{t^*} S(x)$, and constant factor is added to the reward gradient called on the denoiser.

$D_{t^{*}}$ means denoiser evaluated at time $t^{*}$.

So for an "inner glass flow" fix a time $t^{*}$ to evalute the reward gradient, though the image latent to the denoiser for the the reward gradient varies.

* time step $t^{*}$ can be chosen to be the middle of the inner glass flow for now, is a hyperparameter controlling guidance.

$$D_{\mu(s), \Sigma(s)}^r (x) = D_{\mu(s), \Sigma(s)}(x) + \beta_{t^*} \frac{\sigma_{t^*}^2}{\alpha_{t^*}} \nabla_y r(D_{t^*}(y)) |_{y= \alpha_{t^*} S(x)}$$

So there's two calls to the glass denoiser, and a reward model rm.

lhs = glass_denoiser(x)

rhs = glass_denoiser(y,t*)

reward_guided_glass_denoiser = lhs + torch.autograd.grad(rm(rhs),y)

The syntax is like this

```
with torch.enable_grad():
    y = y.detach().requires_grad_(True)
    r = rm(D_tstar(y))
    grad = torch.autograd.grad(r, y)[0]
```
